1 Read the earthquakes.csv file and save its content into a variable. Then use bar chart to visualize magnitude values being 8.2 or over so that date values will be presented in x-axle and magnitude in y-axle.

```python
import pandas as pd
```

```python
df_1 = pd.read_csv('./data_files/earthquakes.csv')

data = df_1[(df_1["Magnitude"] >= 8.2)]

data.plot(kind="bar", x="Date", y="Magnitude", figsize=(25,10))
```

3 Read the gamedata.json file to a new variable. Visualize the data with pie chart so that share count will be counted for each Steam rating (in other words how many percentage out of all games have received the rating of *Very Positive*).

```python
df_2 = pd.read_json("./data_files/gamedata.json") # Toivottavasti ymmrsin tehtvn oikein, kauan aikaa se ainakin vei :D

counts = df_2["steamRatingText"].value_counts()

counts

label_list = [str(round(100 *(i / sum(counts)),1)) + ' %' for i in counts]

counts.plot(kind="pie", y="steamRatingText", labels=label_list, figsize=(10,10)).legend(counts.index)
```
