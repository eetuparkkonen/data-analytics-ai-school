- Use *gamedata.json* file from data_files directory in this exercise. **Important:** Edit the original DataFrame throughout this whole notebook!

1 Read the first 10 rows from the gamedata.json file into the DataFrame. Then read the last five rows from the same file into another DataFrame. Combine these two into one DataFrame.

```python
import pandas as pd
```

```python
df = pd.read_json("./data_files/gamedata.json")

df_first = df.head(10)

df_last = df.tail(5)

combined_df = pd.concat([df_first, df_last]).reset_index(drop=True)

combined_df
```

2 Set *steamRatingPercent* to NaN for all rows having *steamRatingCount* below 1000.

```python

combined_df.loc[combined_df["steamRatingCount"] < 1000, "steamRatingPercent"] = "NaN"
combined_df

```

3 Fill all NaN values in *SteamRatingPercent* column by increasing the *metacriticScore* from the same row with 7 %.

```python
combined_df.loc[combined_df["steamRatingPercent"] == "NaN", "steamRatingPercent"] = combined_df["metacriticScore"] * 1.07

combined_df
```

4 Group the data by *steamRatingText* column so that averages of the following columns will be calculated for each text value: *steamRatingPercent*, *metacriticScore*, *normalPrice* and *salePrice*. Do not show any other columns in the resulting DataFrame.

```python
cols = [0, 1, 2, 3, 4, 5, 8, 9, 13, 14, 15, 16, 17, 18]

combined_df.drop(combined_df.columns[cols],axis=1,inplace=True)
combined_df["steamRatingPercent"] = combined_df["steamRatingPercent"].astype(float) #jostain syystä ei ollut float? Originaalissa oli int valmiina, niin ei siinä tarvinnut tehdä tätä
combined_df = combined_df.groupby("steamRatingText").mean().round(1)
combined_df
```

5 Load the original game_data.json file content into a new DataFrame and then filter the data with the same manner as in previous task. Combine these two DataFrames together and change the column names of the original to contain text *ORIGINAL*. 

```python
original_df = pd.read_json("./data_files/gamedata.json")
original_df.drop(original_df.columns[cols], axis=1, inplace=True)
original_df = original_df.rename(columns={"salePrice":"ORIGINALsalePrice", "normalPrice":"ORIGINALnormalPrice", "metacriticScore":"ORIGINALmetacriticScore","steamRatingPercent":"ORIGINALsteamRatingPercent"})
original_df = original_df.groupby("steamRatingText").mean().round(1)
original_df

df_orig_comb = pd.concat([combined_df, original_df]).reset_index()

df_orig_comb
```
