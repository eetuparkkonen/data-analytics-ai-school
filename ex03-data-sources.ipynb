1 Use gamedata.json for this task. This file contains information of games sold through Steam. Parse out the following information from the data:
- TOP 3 highest metacritic score. Present results using the following format: *Title* has metacritic score of *Score* (for example)
- Games with discount price being 90 % or more. Present results using the following format: *Title* | discount: *Savings* (for example Metal Gear Solid V: Ground Zeroes | score 90.090090)
- Games having metacritic score higher than steam score. Present results using the following format: *Title* has metacritic score of *MetacriticScore* and steam score of *SteamRatingPercent*

```python
import json
```

```python
with open("./data_files/gamedata.json") as data:
    lst = []
    get_title = []
    number = []
    data = json.load(data)
    
    for i in data:
        lst.append(i)
    
for i in lst:
    get_title.append(i['title'])
    number.append(i['metacriticScore'])

top = dict()
        
for i in range(len(number)):
    top[get_title[i]] = number[i]

sorted_lst = sorted(top.items(), key=lambda x: x[1], reverse=True)[:3]

print('{} has metacritic score of: {}'.format(sorted_lst[0][0], sorted_lst[0][1]))
print('{} has metacritic score of: {}'.format(sorted_lst[1][0], sorted_lst[1][1]))
print('{} has metacritic score of: {}'.format(sorted_lst[2][0], sorted_lst[2][1]))
```

```python
with open("./data_files/gamedata.json") as data:
    data = json.load(data)
    discountList = []
    
    for i in data:
        if float(i["savings"]) >= 90:
            discountList.append(i)
    for i in discountList:
        print('{} | discount: {}'.format(i["title"], i['savings']))
```

```python
with open("./data_files/gamedata.json") as data:
    data = json.load(data)
    listed = []
    
    for i in data:
        if float(i['metacriticScore']) > float(i['steamRatingPercent']):
            listed.append(i)
            
    for i in listed:
        print('{} has metacritic score of {} and steam score of {}'.format(i['title'], i['metacriticScore'], i['steamRatingPercent']))
        
```

2 Use earthquakes.csv for this task. This file contains information about earthquakes recorded between 1965 and 2016. Earthquake magnitude value describes how strong the earthquake is. Magnitude information can be categorized like presented in the table below (*Source: http://www.geo.mtu.edu/UPSeis/magnitude.html*).

| Magnitude      | Class | Effects |
|----------------|-------|---------|
| 2.5 or less    | Minor | Usually not felt, but can be recorded by seismograph. |
| 2.5 to 5.4     | Light | Often felt, but only causes minor damage. |
| 5.5 to 6.0     | Moderate | Slight damage to buildings and other structures. |
| 6.1 to 6.9     | Strong | May cause a lot of damage in very populated areas. |
| 7.0 to 7.9     | Major | Major earthquake. Serious damage. |
| 8.0 or greater | Great | Great earthquake. Can totally destroy communities near the epicenter. |

Count how many earthquakes have occurred in each class.

```python
import csv
```

```python
with open('./data_files/earthquakes.csv') as data:
    data = csv.reader(data)
    arr = []
    minor = 0
    light = 0
    moderate = 0
    strong = 0
    major = 0
    great = 0
    
    for i in data:
        arr.append(i)
    
    arr.pop(0)
    
        

    for i in arr:
        if float(i[8]) <= 2.5:
            minor += 1
        elif float(i[8]) >  2.5 and float(i[8]) < 5.5:
            light += 1
        elif float(i[8]) >= 5.5 and float(i[8]) <= 6.0:
            moderate += 1
        elif float(i[8]) > 6 and float(i[8]) <= 6.9:
            strong += 1
        elif float(i[8]) >= 7.0 and float(i[8]) <= 7.9:
            major += 1
        else:
            great += 1
            
            
            
print('Amount of earthquakes happened: minor quakes: {}, light quakes: {}, moderate quakes: {}, strong quakes: {}, major quakes: {}, great quakes: {}'.format(minor, light, moderate, strong, major, great))

            
            
```

3 Use netflix_titles.xml for this task. This file contains information about Netflix movies and TV shows. **Important:** Movies have duration presented in minutes while TV shows have duration presented in amount of seasons! Parse out the following information from the data:
- Movies released in 2017
- TV show and movie amount (present both counts in separate lines)
- Movies with a length between 15 and 20 minutes (values 15 and 20 included)

```python
import xml.etree.ElementTree as e
```

```python
tree = e.parse('./data_files/netflix_titles.xml')
root = tree.getroot()

lst = []
lst2 = []
for i in root:
    obj = {}
    obj2 = {}
    for j in i:
        if j.tag == 'title':
            obj[j.tag] = j.text
        if j.tag == 'release_year':
            obj2[j.tag] = j.text
    lst.append(obj)
    lst2.append(obj2)
    
for i in range(len(lst2)):
    if lst2[i]['release_year'] == '2017':
        print("movie title: {}, release year: {}".format(lst[i]['title'], lst2[i]['release_year']))
   

```

```python
tree = e.parse('./data_files/netflix_titles.xml')
root = tree.getroot()

tvshow = 0
movie = 0

for i in root:
    for j in i:
        if j.tag == "type":
            if j.text == "Movie":
                movie += 1
            else:
                tvshow += 1
            
print("Movie amount:", movie)
print("Tv show amount:", tvshow)
```

4 Use the following Rest API for this task: https://tie.digitraffic.fi/api/v1/data/weather-data. Calculate the average for air temperature (ILMA) and humidity (ILMAN_KOSTEUS) values using two decimals.

```python
import requests
```

```python
url = "https://tie.digitraffic.fi/api/v1/data/weather-data"

req = requests.get(url=url)

data = req.json()

data_ilma = []
sensor = []
        
    
# ILMAN KESKILÄMPÖTILA
        
for i in range(len(data["weatherStations"])):
    if data["weatherStations"][i]["sensorValues"]:
        for j in range(len(data["weatherStations"][i]["sensorValues"])):
            if data["weatherStations"][i]["sensorValues"][j]["name"] == "ILMA":
                data_ilma.append(data["weatherStations"][i]["sensorValues"][j])
    

for i in range(len(data_ilma)):
    sensor.append(data_ilma[i]["sensorValue"])

    
res1 = sum(sensor)

avgIlma = round(res1 / len(sensor), 2) 


print("Average air temperature is {} celsius degrees".format(avgIlma))

```

```python
#ILMANKOSTEUS

url = "https://tie.digitraffic.fi/api/v1/data/weather-data"

req = requests.get(url=url)

data = req.json()

data_kosteus = []

kosteus = []


for i in range(len(data["weatherStations"])):
    if data["weatherStations"][i]["sensorValues"]:
        for j in range(len(data["weatherStations"][i]["sensorValues"])):
            if data["weatherStations"][i]["sensorValues"][j]["name"] == "ILMAN_KOSTEUS":
                data_kosteus.append(data["weatherStations"][i]["sensorValues"][j])
                
for i in range(len(data_kosteus)):
    kosteus.append(data_kosteus[i]["sensorValue"])
    
res2 = sum(kosteus)

avgKosteus = round(res2 / len(kosteus), 2)

print("Average air humidity is {}%".format(avgKosteus))
```
